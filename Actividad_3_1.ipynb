```python
#Carga desde un archivo .xslx sin indice
df = pd.read_csv('Actividad2.1_Sin_Nulos.csv')
data=df.drop(['Unnamed: 0'],axis=1)
data.head()
```

```python
#Identificar valores nulos por columnas
valores_nulos = df.isnull().sum() #desglosado
#valores_nulos = data2.isnull().sum().sum() #Completo
valores_nulos
```

#1.-Primero

```python
#Realizamos diagrama de bigote
fig = plt.figure(figsize=(10,6))
plt.boxplot(data["ventas_precios_corrientes"])
plt.title("Outliers de ventas_precios_corrientes")
plt.show()
```

```python
#Mtodo aplicando Cuartiles. Encuentro cuartiles 0.25 y 0.75
y=data['ventas_precios_corrientes']

percentile25=y.quantile(0.25) #Q1
percentile75=y.quantile(0.75) #Q3
iqr= percentile75 - percentile25

Limite_Superior_iqr= percentile75 + 1.5*iqr
Limite_Inferior_iqr= percentile25 - 1.5*iqr
print('Limite superior permitido', Limite_Superior_iqr)
print('Limite superior permitido', Limite_Inferior_iqr)
```

```python
#Encontramos outliers
outliers_iqr = data[(y>Limite_Superior_iqr)|(y<Limite_Inferior_iqr)]
outliers_iqr
```

```python
#Obtenemos datos limpios
data_clean_iqr1 = data[(y<=Limite_Superior_iqr)&(y>=Limite_Inferior_iqr)]
data_clean_iqr1
```

```python
#Realizamos diagrama de bigote
fig = plt.figure(figsize=(10,6))
plt.boxplot(data_clean_iqr1["ventas_precios_corrientes"])
plt.title("Outliers de ventas_precios_corrientes")
plt.show()
```
